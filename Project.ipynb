```python
# Create a countplot
ax = sns.countplot(x="Attrition", data=hr)
plt.title("Employee Attrition Distribution")

# Function to add labels on bars
def add_labels(ax):
    for p in ax.patches:
        width = p.get_width()
        height = p.get_height()
        x, y = p.get_xy()
        ax.annotate(f'{height}', (x + width/2, y + height), ha='center', va='bottom')

# Add labels to the countplot
add_labels(ax)

plt.show()
```

```python


categorical_features = [
    "BusinessTravel", "Department", "EducationField",
    "Gender", "JobRole", "MaritalStatus", "OverTime"
]

# Function to add labels on bars
def add_labels(ax):
    for p in ax.patches:
        width = p.get_width()
        height = p.get_height()
        x, y = p.get_xy()
        ax.annotate(f'{height}', (x + width/2, y + height), ha='center', va='bottom')

for feature in categorical_features:
    plt.figure(figsize=(10, 4))
    ax = sns.countplot(x=feature, hue="Attrition", data=hr)
    plt.title(f"Attrition by {feature}")
    add_labels(ax)
    plt.show()
```

```python


categorical_features = [
    "BusinessTravel", "Department", "EducationField",
    "Gender", "JobRole", "MaritalStatus", "OverTime"
]

# Function to add labels on bars
def add_labels(ax):
    for p in ax.patches:
        width = p.get_width()
        height = p.get_height()
        x, y = p.get_xy()
        ax.annotate(f'{height}', (x + width/2, y + height), ha='center', va='bottom')

# Function to create a pie chart
def create_pie_chart(data, feature, ax):
    attrition_by_feature = data.groupby([feature, "Attrition"]).size().unstack(fill_value=0)
    attrition_percent = attrition_by_feature.div(attrition_by_feature.sum(axis=1), axis=0)
    attrition_percent.plot(kind='pie', subplots=True, ax=ax, autopct='%.1f%%', legend=False)

# Function to create a stacked bar chart
def create_stacked_bar_chart(data, feature, ax):
    attrition_by_feature = data.groupby([feature, "Attrition"]).size().unstack(fill_value=0)
    attrition_percent = attrition_by_feature.div(attrition_by_feature.sum(axis=1), axis=0)
    attrition_percent.plot(kind='bar', stacked=True, ax=ax)
    ax.legend(title='Attrition')

for feature in categorical_features:
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle(f"Attrition by {feature}")

    # Bar plot
    ax = sns.countplot(x=feature, hue="Attrition", data=hr, ax=axes[0])
    add_labels(ax)
    axes[0].set_title(f"{feature} - Bar Plot")

    # Pie chart
    create_pie_chart(hr, feature, axes[1])
    axes[1].set_title(f"{feature} - Pie Chart")

    # Stacked bar chart
    create_stacked_bar_chart(hr, feature, axes[2])
    axes[2].set_title(f"{feature} - Stacked Bar Chart")

    plt.show()
```
